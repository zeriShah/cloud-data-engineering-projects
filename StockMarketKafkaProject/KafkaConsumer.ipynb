```python
# Ensure S3 is connected with correct config
import s3fs
import json
from kafka import KafkaConsumer
from json import loads

s3 = s3fs.S3FileSystem(
    key="YOUR_AWS_ACCESS_KEY_ID",
    secret="YOUR_AWS_SECRET_ACCESS_KEY",
    client_kwargs={'region_name': 'us-east-1'}
)

# Create fresh consumer
consumer = KafkaConsumer(
    'demo_test',
     bootstrap_servers=['98.93.74.48:9092'],
    value_deserializer=lambda x: loads(x.decode('utf-8')))

print("Starting to consume all messages from Kafka and write to S3...")
print("Press Ctrl+C to stop\n")

for count, i in enumerate(consumer):
    with s3.open(f"s3://s3-kafka-data/stock_market_{count}.json", 'w') as file:
        json.dump(i.value, file)
    
    # Print progress every 100 messages
    if count % 100 == 0:
        print(f"Processed {count} messages...")

print(f"Completed! Total messages written to S3: {count + 1}")
```
