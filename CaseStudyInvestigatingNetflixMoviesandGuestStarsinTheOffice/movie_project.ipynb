### Movies ETL Project (API Based)
A basic Extract, Transform and Load (ETL) pipeline using API, pandas and SQLite

```python
import requests 
import pandas as pd 
import sqlite3
from datetime import datetime
```

###  Step 0: Maintaining a Log File

```python
def log_progress(message):
    with open("./logs/movies_logs", 'a') as f:
        f.write(f'{datetime.now()} : {message}\n')
```

### Step 1: Extratct (from API)

```python
def extract(api_url):
   """extrat movies data from API"""
   response = requests.get(api_url)
   data = response.json()   # yahan JSON aayega
   df = pd.DataFrame(data)
   log_progress("data extraction from API Completed. Initiating transformtion")
   return df
```

### Step 2: tranform Data

```python
def transform(df):
    """Convert rating from 10-point to 5-point scale"""
    df['Rating_5']= round(df['rating']/2,1)
    log_progress("data transformation compelte")
    return df
```

## Step 3: Load 


```python
### csv 
def load_to_csv(df,output_path):
    df.to_csv(output_path,index=False)
    log_progress("data saved to CSV file")

```

```python
## sqlite

def load_to_db(df, sql_conn, table_name):
    df.to_sql(table_name, sql_conn,if_exist='replace', index=False)
    log_progress("data loaded to database")
```

### step 04: run query

```python
def run_query(query_statement, sql_conn):
    print(pd.read_sql(query_statement, sql_conn))
    log_progress(f"Executed query: {query_statement}")
```

```python
API_URL = "https://api.tvmaze.com/shows" # sample API

#extract
movies_df = extract(API_URL)

#transform
movies_df= transform(movies_df)

#load
load_to_csv(movies_df, "./movies_api_data.csv")

conn = sqlite3.connect("MoviesAPI_DB.db")
load_to_db(movies_df, conn, "MoviesAPI") # Run Queries run_query("SELECT * FROM MoviesAPI", conn) run_query("SELECT title, Rating_5 FROM MoviesAPI WHERE Rating_5 >= 4.5", conn) conn.close()
```
