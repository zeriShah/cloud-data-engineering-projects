```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# Setup WebDriver
options = Options()
options.add_argument("--start-maximized")  
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
driver = webdriver.Chrome(options=options)

url = 'https://coinmarketcap.com/'
driver.get(url)
time.sleep(3)  

# Slowly scroll to the bottom of the page
scroll_pause_time = 0.5 
scroll_step = 300  # pixels per scroll
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    for i in range(0, last_height, scroll_step):
        driver.execute_script(f"window.scrollTo(0, {i});")
        time.sleep(scroll_pause_time)
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  
    last_height = new_height

print("✅ Finished scrolling")



soup = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()

table = soup.find('table', {'class': 'cmc-table'})

rows = table.find_all('tr')
print(f"✅ Found {len(rows)-1} rows (excluding header)")

crypto_data = []

for row in rows[1:]: 
    cols = row.find_all('td')
    if len(cols) >= 10:
        rank = cols[1].text.strip()
        name = cols[2].text.strip()
        price = cols[3].text.strip()
        one_hour_change = cols[4].text.strip()
        twenty_four_hour_change = cols[5].text.strip()
        seven_day_change = cols[6].text.strip()
        market_cap = cols[7].text.strip()
        hr_volume = cols[8].text.strip()
        circulating_supply = cols[9].text.strip()

        crypto_data.append([
            rank, name, price, one_hour_change,
            twenty_four_hour_change, seven_day_change,
            market_cap, hr_volume, circulating_supply
        ])

for coin in crypto_data:
    print(coin)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# =======================
# WebDriver Setup
# =======================
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=options)

all_crypto_data = []

# =======================
# Pagination Loop
# =======================
for page in range(1, 4):
    url = f"https://coinmarketcap.com/?page={page}"
    driver.get(url)
    time.sleep(3)

    print(f"📄 Page {page} loaded, step scrolling...")

    # =======================
    # STEP SCROLLING (400px)
    # =======================
    scroll_step = 400
    scroll_pause = 1.2

    current_position = 0
    page_height = driver.execute_script("return document.body.scrollHeight")

    while current_position < page_height:
        driver.execute_script(f"window.scrollTo(0, {current_position});")
        time.sleep(scroll_pause)

        current_position += scroll_step
        page_height = driver.execute_script("return document.body.scrollHeight")

    print("✅ Scrolling completed")

    # =======================
    # Parse Page
    # =======================
    soup = BeautifulSoup(driver.page_source, "html.parser")
    table = soup.find("table")

    if not table:
        print("❌ Table not found on page", page)
        continue

    rows = table.find_all("tr")

    for row in rows[1:]:
        cols = row.find_all("td")
        if len(cols) >= 10:
            rank = cols[1].text.strip()
            name = cols[2].text.strip()
            price = cols[3].text.strip()
            one_hour = cols[4].text.strip()
            twenty_four = cols[5].text.strip()
            seven_day = cols[6].text.strip()
            market_cap = cols[7].text.strip()
            volume = cols[8].text.strip()
            supply = cols[9].text.strip()

            all_crypto_data.append([
                page, rank, name, price, one_hour,
                twenty_four, seven_day, market_cap,
                volume, supply
            ])

# =======================
# Close Browser
# =======================
driver.quit()

print(f"\n🔥 TOTAL COINS COLLECTED: {len(all_crypto_data)}\n")

# Preview first 10 rows
for coin in all_crypto_data[:10]:
    print(coin)
```

```python
# db loading


from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pyodbc

# ==================================================
# DATABASE CONFIG (SAME STYLE AS YOUR OTHER PROJECT)
# ==================================================
server = "DESKTOP-K4EF3Q0"
username = "sa"
password = "shah123"
db_name = "CoinDB"

# ---------- Connect to MASTER ----------
conn = pyodbc.connect(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};DATABASE=master;"
    f"UID={username};PWD={password}",
    autocommit=True
)
cursor = conn.cursor()
print("✅ Connected to SQL Server (master)")

# ---------- Create Database ----------
cursor.execute(f"SELECT name FROM sys.databases WHERE name='{db_name}'")
if not cursor.fetchone():
    cursor.execute(f"CREATE DATABASE {db_name}")
    print(f"✅ Database '{db_name}' created")
else:
    print(f"ℹ Database '{db_name}' already exists")

conn.close()

# ---------- Connect to CoinDB ----------
conn = pyodbc.connect(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};DATABASE={db_name};"
    f"UID={username};PWD={password}",
    autocommit=True
)
cursor = conn.cursor()
cursor.fast_executemany = True
print(f"✅ Connected to {db_name}")

# ---------- Create Table ----------
cursor.execute("""
IF OBJECT_ID('dbo.CryptoData', 'U') IS NULL
CREATE TABLE dbo.CryptoData (
    id INT IDENTITY(1,1) PRIMARY KEY,
    page_no INT,
    rank_no NVARCHAR(50),
    name NVARCHAR(255),
    price NVARCHAR(50),
    one_hour_change NVARCHAR(50),
    twenty_four_hour_change NVARCHAR(50),
    seven_day_change NVARCHAR(50),
    market_cap NVARCHAR(100),
    volume_24h NVARCHAR(100),
    circulating_supply NVARCHAR(100),
    created_at DATETIME DEFAULT GETDATE()
)
""")
print("✅ Table CryptoData ready")

# ==================================================
# SELENIUM SETUP
# ==================================================
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=options)

# ==================================================
# SCRAPING (PAGINATION + STEP SCROLLING)
# ==================================================
for page in range(1, 4):
    print(f"📄 Scraping page {page}")
    driver.get(f"https://coinmarketcap.com/?page={page}")
    time.sleep(3)

    # ---- STEP SCROLLING (400px) ----
    scroll_step = 400
    scroll_pause = 1.2
    pos = 0
    height = driver.execute_script("return document.body.scrollHeight")

    while pos < height:
        driver.execute_script(f"window.scrollTo(0, {pos});")
        time.sleep(scroll_pause)
        pos += scroll_step
        height = driver.execute_script("return document.body.scrollHeight")

    # ---- Parse Page ----
    soup = BeautifulSoup(driver.page_source, "html.parser")
    table = soup.find("table")
    if not table:
        print("❌ Table not found")
        continue

    rows = table.find_all("tr")

    for row in rows[1:]:
        cols = row.find_all("td")
        if len(cols) >= 10:
            cursor.execute("""
                INSERT INTO dbo.CryptoData (
                    page_no, rank_no, name, price,
                    one_hour_change, twenty_four_hour_change,
                    seven_day_change, market_cap,
                    volume_24h, circulating_supply
                ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            """,
            page,
            cols[1].text.strip(),
            cols[2].text.strip(),
            cols[3].text.strip(),
            cols[4].text.strip(),
            cols[5].text.strip(),
            cols[6].text.strip(),
            cols[7].text.strip(),
            cols[8].text.strip(),
            cols[9].text.strip()
            )

    conn.commit()
    print(f"✅ Page {page} inserted into DB")

# ==================================================
# CLEANUP
# ==================================================
driver.quit()
conn.close()

print("\n🔥 ALL DATA SUCCESSFULLY SAVED IN SQL SERVER (CoinDB)")
```
